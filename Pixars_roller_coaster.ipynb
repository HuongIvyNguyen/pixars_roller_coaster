```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pixar_movies = pd.read_csv("PixarMovies.csv")
nrow = pixar_movies.shape[0]
ncol = pixar_movies.shape[1]

print('\nThe number of rows for this dataset is %d' %nrow + ' and the number of columns is %d' %ncol)
```

```python
#Elimate all percentage sign from the Domestic % and International % columns
pixar_movies["Domestic %"] = pixar_movies["Domestic %"].str.rstrip("%").astype("float")
pixar_movies["International %"] = pixar_movies["International %"].str.rstrip("%").astype("float")
```

```python
#Converting IMDB score to 100 scale
pixar_movies["IMDB Score"] = pixar_movies["IMDB Score"]*10
```

```python
pixar_movies
```

```python
filtered_pixar = pixar_movies.dropna()
```

```python
filtered_pixar
```

```python
#Make the Movie column the index column for both pixar_movies and filtered_pixar dataframes
pixar_movies.set_index("Movie", inplace = True)
filtered_pixar.set_index("Movie", inplace = True)
```

```python
critic_reviews = filtered_pixar[['RT Score','IMDB Score', 'Metacritic Score']]
critic_reviews.plot(figsize=(10,6))
plt.show()

critic_reviews.plot(kind = 'box', figsize=(9,5))
plt.show()
```

```python
revenue_proportion = filtered_pixar[['Domestic %', 'International %']]
revenue_proportion.plot(kind ='bar', stacked = True)
```
